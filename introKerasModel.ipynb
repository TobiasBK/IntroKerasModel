```
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```
train_labels = []
train_samples = []
```

Example Data (fake data for now, real data later):


*   An experimental drug was tested on individuals from ages 13 to 100 in a clinical trial
*   The trial had 2100 participants. Half were under 65 y/o, half were older than 65 y/o
*   Around 95% or pts > 65 y/o had side effects.
*   Around 96% of pts < 65 y/o had NO side effects. 






```
#creating the data set
for i in range (50):
  # The -5% of younger individuals who did experience side effects
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(1) #1 represent pts that DID have side effects. 0 would = no sideeffects

  #The ~5% of older people who did NOT have side effects
  random_older = randint(65,100)
  train_samples.append(random_older)
  train_labels.append(0)

for i in range(1000):
  # 95% of younger people who did not have side effects
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(0)

  # 95% of older people who did experience side effects
  random_older = randint(65,100)
  train_samples.append(random_older)
  train_labels.append(1)
```

```
#visualize samples(ages)
for i in train_samples:
  print(i)
```

```
#vizualize side effects vs no side effects
for i in train_labels:
  print(i)
```

```
#now need to process the data so that the data is in 
#the format that the fit fx expects
#taking lists and making np arrays
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples) #gets ride of accidental order
```

```
#Here, we're normalizing the data so that the training of the 
#NN wiil become faster and more efficient
#creates featrue range from 0 to 1:
scaler = MinMaxScaler(feature_range=(0,1))
#using above line + this to rescale the data from 13-100 down to 0 to 1. Needs to be reshpaed to transform:
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```
for i in scaled_train_samples:
  print(i)
```

```
#used fro building the model:

from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
#used later for training the model:
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```
#only run this if using a GPU:
# physical_devices = tf.config.experimental.list_physical_devices('GPU')
# print("Num GPUs Available: ", len(physical_devices))
# tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

So the model is a sequential model. It's the simpliest model to build using keras and TF. So, you initialize the model and then you pass in a list of layers. The first dense layer is actually the second layer. It is, rather, the first hidden layer. The input layer is not obvious b/c it is not obviously defined. In other words, the input data creates the input layer. Dense (a densely or fully connected layer) is a common type of layer. 

The model understands the shape of the input data from the input_shape function passed in the second layer. The units fx is how many nodes/neurons will be used. The number 16 is a guess of what wrks best. 


Wiki: "In artificial neural networks, the activation function of a node defines the output of that node given an input or set of inputs. A standard integrated circuit can be seen as a digital network of activation functions that can be "ON" (1) or "OFF" (0), depending on input."

 The last layer is the output layer. Either the pt. did or didn't (1 or 0) suffer side effects. The softmax fx will give probabilities for each output class (1 or 0). 

```
model = Sequential([
                    Dense(units=16, input_shape=(1,), activation='relu'),
                    Dense(units=32, activation='relu'),
                    Dense(units=2, activation='softmax')
])
```

```
#See a print out:
model.summary()
```

```
#prepare model for training:
#Specify what optimizer to use. Here using Adam(very common) w learning rate of 0.0001
#Metrics is a common way to eval model performance.
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

shuffle = default is True. Removes any accidental order. So, the model dosn't learn from ordered data.

##Now, Creating a validation set. So separate the data so you have training data and then validation data.


The validation set is just to see how weel the model can geranlize on data not included in training data.
This will allow us to see if overfitting occurs. Overfitting is when a model learns the specifics of the training data very well, but can't generalize on data it hasn't seen before

So, if the model gets excellent results for the training set but poor results for the validation set, you can conclude that you have an overfitting problem. 

Now, let's create a validation set with a Keras sequential model.
There are two ways:
1. Create a separate validation set from training set and then pass the validation set to the model using the fit function.
2. OR you can get Keras to create a validation set for you, this way saves time.

```
#So, this time including a validation set.
#Pick a number btw 0 and 1.
#Here we are picking 10% of the training samples will be held out for the validation set.
#Important: The validation_split will happen before the data is shuffled,
#the validation set will be the last percentage of your training data, therefore it is
#possible to take a bunch of unshuffled data skewing results.
#To help mitigate this, make sure data is SHUFFLED before being passed to the fit function!! Especiall w validation_split in use.
model.fit(x = scaled_train_samples, y=train_labels, validation_split=0.1, batch_size=10, epochs=30, shuffle=True, verbose=2)
```

So, now we compare the accuracy(.9487) to the val_accuracy(.9429). In this case they are very similar especailly by epoch 30. This is good. It indicates there is likely no overfitting. If, however, we saw the val_accuracy lagging the training accuracy then we could conclude over fitting. 
We can also compare training loss(.2308) and val_loss(.2270) and see that they are similar. This is also good.

## Confusion Matrix

We will use a confusion matrix to visualize results following inference

```
 %matplotlib inline
 from sklearn.metrics import confusion_matrix
 import itertools
 import matplotlib.pyplot as plt
```

```
cm = confusion_matrix(y_true=test_labels, y_pred=rounded_predictions)
```

```
#copied directly from sklearn website:
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
  """
  This function prints and plots the confusion matrix.
  Normalization can be applied by setting 'normalize=True'.
  """
  plt.imshow(cm, interpolation='nearest', cmap=cmap)
  plt.title(title)
  plt.colorbar()
  tick_marks = np.arange(len(classes))
  plt.xticks(tick_marks, classes, rotation=45)
  plt.yticks(tick_marks, classes)

  if normalize:
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    print("Normalized confusion matrix")
  else:
    print('Confusion matrix, without normalization')

  print(cm)

  thresh = cm.max() / 2.
  for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, cm[i, j],
             horizontalalignment="center",
             color="white" if cm[i, j] > thresh else "black")
    
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```
cm_plot_labels = ['no_side_effects', 'had_side_effects']
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title='Confusion Matrix')
```

Interpretation of above output. Read from top left corner to bottom right corner. So, the model stated no side effects 10 times when the pt had side effects. And the model stated pt had side effects 18 times when the patient really had no side effects.

So, 392 correct predictions out of 420 total patients. So, a ~93% accuracy rate on the test set. This is similar to what we saw for out validation accuracy during the models training.

So, a confusion matrix is a great tool to use to confirm and visualize how well the model is doing at its predictions. Plus, it can show you which areas the model may need to be improved.

## Build and Train CNN

Using the Keras Sequential Model

```
model = Sequential([
                    Conv2D(filters=32, kernel_size=(3,3), activation='relu', padding='same', input_shape=(224,224,3)),
                    MaxPool2D(pool_size=(2,2), strides=2),
                    Conv2D(filters=64, kernel_size=(3,3), activation='relu', padding='same'),
                    MaxPool2D(pool_size=(2,2), strides=2),
                    Flatten(),
                    Dense(units=2, activation='softmax')
])
```

```
model.summary()
```

B/c we have only two outputs from our model (cat or dog) it is possible to use binary_crossentropy and not categorical_crossentropy. If we did that we would need 1 output node from our model instead of 2 AND we would need to follow that with the sigmoid function rather than the softmax activation function. Can use either or in this circumstance. Crossentropy typically used when you have more than 2 classes. But b/c the results are the same, one less thing to change. 

```
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```
model.fit(x=train_batches, validation_data=valid_batches, epochs=10, verbose=2)
```

With the output from above: The accuracy rate is 100% by the 8th epoch. This is good. But the validation accuracy is 65%. This is bad. This difference indicates overfitting. If this was a model that we cared about and wanted to deploy to production, then we would need to fix this Now. But we won't do that now, we are going to use this data for later to learn fine tuning.

The pre-trained model is called: VGG16

Link to the original VGG16 paper: https://arxiv.org/pdf/1409.1556.pdf

Let's talk about preporcessing a little bit. We kinda skipped over it above. If we look under ther ARCHITECTURE section in the VGG16 paper we find that "The only preprocessing we do is subtracting the mean RGB value, computed on the training set, from each pixel."
This means that they computed the mean RED value pixel from all of the training data. Then across each image in the training set, they subtracted the mean value from the red value in each image. Then repeated for green and blue pixel values. So, b/c this is the way the VGG16 model was originally trained, it means that when new data is passed to the model, it needs to be trained in the same exact way.

Thankfully, Keras already has build in functions for popular models like VGG16 that can do a lot of the preprocessing work. 

## Build Fine-tuned VGG16 model

```
#download model
vgg16_model = tf.keras.applications.vgg16.VGG16()
```

```
vgg16_model.summary()
```

So, the above output shows that the vgg16 model is very complex.
Our goal at first, is to change the ouput predictions from 1000 to just 2 (cat or dog)

```
type(vgg16_model)
```

The above output shows that the model is called Model and is from the Keras functional API. Previously, we were working w Sequential models. Later we will discuss the functional API in depth (bit more sophisticated). 

```
#We won't use functional API yet, so first we will convert the
#Model into a Sequential model by:
#looping thru each layer except for the last one (-1) and add each layer into 
#our model. 
model = Sequential()
for layer in vgg16_model.layers[:-1]:
  model.add(layer)
```

```
model.summary()
```

The output above: shows that this new model summary compared to the older summary, shows that the models are the same EXCEPT that the predictions layer (-1). The fully connected layer (fc) is now the last layer.

```
#now going to iterate over all of our layers
#and set them to not be trainable
#this freezes the trainable parameters(weights and biases) from all the layers
#in the model. This means they won't be RETRAINED when we go thru the trianing 
#process for cats and dogs
#vgg16 has already learned the features in its orginial training, tus we freeze the training here.
for layer in model.layers:
  layer.trainable = False
```

```
#adding our own output layer to replace the one we removed
model.add(Dense(units=2, activation='softmax')) 
```

```
model.summary()
```

So, we will only train the last layer(the new one) b/c everything is laready pretrained. Now we only have 8,000 parameters

###Train the fine-tuned VGG16 model

```
model.compile(optimizer=Adam(learning_rate=0.0001), loss="categorical_crossentropy", metrics=['accuracy'])
```

```
model.fit(x=train_batches, validation_data=valid_batches, epochs=5, verbose=2)
```

So, after just 5 epochs, accuracy iss 99% and validation accuracy is 98%. This is very good. 
This isn't surprising, of course, be/c vgg16 has already been trained on images of cats and dogs.

So, comparing vgg16 to the CNN model we built from scratch, it is clear that the vgg16 model generalized much better on the cat and dog data.
So, since the accuracy is really high on the validation set we will now use the vgg16 model to predict on the test data

### Predict using fine-tuned VGG16 Model

### DATA Augmentation on images in TF

Data augmentation happens when we create new data by making changes(augmentations) to existing data.
For images, examples of this would be flipping the image vertically or horizaontally, rotation, color changes, etc. 

WE use this when we don't have access to enough data. And if our training set is too small we can have problems like overfitting. And so, augmenting data can grow our training set. 

```
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```
